- 입력받은 문자열의 모든 괄호를 뽑아서 올바른 순서대로 배치된 괄호 문자열을 알려주는 프로그램을 작성하시오.
- 단, 입력된 괄호는 개수는 맞지만 짝이 다르다
    - 균형잡힌 괄호 문자열 : ( 와 )의 개수가 같은 경우
    - 올바른 괄호 문자열 : 균형잡힌 괄호 문자열이면서 짝도 모두 맞을 경우
    - 균형잡힌 문자열을 올바른 괄호 문자열로 변환하는 프로그램
    - 균형잡힌 괄호 문자열이 아닌 입력이 들어오는 경우는 없다
- 과정
    1. 입력이 빈 문자열 => 빈 문자열 반환
    2. 문자열 w를 두 균형잡힌 괄호 문자열 u, v로 분리
        - 단, u는 균형잡힌 괄호 문자열로 더 이상 분리할 수 없어야 하며, v는 빈 문자열이 될 수 있음
    3. 수행한 결과 문자열을 u에 이어 붙인 후 반환  
        3-1. 문자열 u가 올바른 괄호 문자열이라면 문자열 v에 대해 1단계부터 다시 수행  
    4. 문자열 u가 올바른 괄호 문자열이 아니라면 아래 과정을 수행  
        4-1. 빈 문자열에 첫 번째 문자로 (를 붙임  
        4-2. 문자열 v에 대해 1단계부터 재귀적으로 수행한 결과 문자열을 이어 붙임  
        4-3. )를 다시 붙임  
        4-4. u의 첫 번째와 마지막 문자를 제거하고, 나머지 문자열의 괄호 방향을 뒤집어서 뒤에 붙임  
        4-5. 생성된 문자열을 반환

```python
# 균형잡힌 괄호 문자열의 인덱스 반환
def balanced_index(p):
    # 왼쪽 괄호의 개수 => 기준
    count = 0
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        else:
            count -= 1
        
        # count가 0이란 뜻은 (와 )의 개수가 같다는 뜻
        # 즉, 처음으로 0이된 구간의 index를 반환한다 => w를 균형잡힌 괄호 문자열 두 개로 나누기 때문
        # 입력되는 문자열은 균형잡힌 괄호 문자열이 아닌 경우가 없으므로 무조건 균형잡힌 괄호 문자열로 다시 나누어진다. (또는 빈문자열)
        if count == 0:
            return i
        
# 올바른 괄호 문자열인지 판단
def check_proper(p):
    # 왼쪽 괄호의 개수
    count = 0
    for i in p:
        if i == '(':
            count += 1
        else:
            # 처음이 왼쪽 괄호가 아닌 경우
            # 즉, 오른쪽 괄호로 문자열이 시작되는 경우에만 count가 0이 된다
            # 이 말은 오른쪽 괄호로 시작되는 문자열 자체가 잘못된 문자열이므로 False를 리턴
            if count == 0:
                return False
            count -= 1
            
    return True

def solution(p):
    
    # 빈 문자열이면 그대로 출력
    answer = ''
    if p == '':
        return answer
    
    index = balanced_index(p)
    u = p[:index + 1]
    v = p[index + 1:]
    
    # 올바른 괄호 문자열이면, v에 대해 함수를 수행한 결과 붙여 반환
    if check_proper(u):
        answer = u + solution(v)
    # 올바른 괄호 문자열이 아니라면 아래의 과정을 수행
    else:
        # answer가 빈 문자열
        answer = '('
        answer += solution(v)
        answer += ')'
        
        # 첫 번째와 마지막 문자열 제거
        u = list(u[1:-1])
        
        # 제거하고 남은 문자열의 괄호를 서로 바꿈
        for i in range(len(u)):
            if u[i] == '(':
                u[i] = ')'
            else:
                u[i] = '('
        answer += "".join(u)
        
        
        
    return answer
```
