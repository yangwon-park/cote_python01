- 뱀이 기어다니면서 사과를 먹는데 먹으면 뱀 길이가 늘어남
- 벽 또는 자기 자신의 몸과 부딪히면 게임이 끝남
- N x N 정사각 보드 위에서 진행 (몇몇 칸에는 사과가 존재)
- 보드 상하좌우 끝에는 벽이 존재
- 시작
    - 위치 : 0,0 (여기엔 사과가 없음)
    - 뱀 길이 : 1
    - 방향 : 오른쪽
- 매 초마다 움직이는 규칙
    - 몸길이를 늘려 머리를 다음 칸에 위치
    - 이동한 칸에 사과가 있다면, 그 칸에 있던 사과가 없어지고 꼬리는 움직이지 않음
    - 이동한 칸에 사과가 없다면, 몸길이를 줄여서 꼬리가 위치한 칸을 비워줌

```python
# 보드의 크기
N = int(input())

# 사과의 개수
K = int(input())

# 맵 정보를 담을 리스트
# 처음 시작이 0,0이 아닌 1,1로 할 것 => N이 아닌 N+1만큼 크기 부여
board = [[0] * (N+1) for _ in range(N+1)]

# 사과가 존재하는 곳은 1로 표시
for _ in range(K):
    a, b = map(int, input().split())
    board[a][b] = 1
    
# 뱀의 방향 변환 횟수
L = int(input())

# 방향 전환 정보를 담을 리스트
direc = []

# 뱀의 방향 변환 정보 (정수 X, 문자 C)
# 게임 시작 시간으로부터 X초가 끝난 뒤에 왼쪽 (C : L) 또는 오른쪽 (C : D)으로 90도 방향 회전시킨다는 뜻
# X <= 10000, X가 증가하는 순으로 주어짐
for _ in range(L):
    X, C = input().split()
    direc.append((int(X), C))
    
    
# 바라보는 방향 전환이 있다? => dx, dy 바로 구현
# 동 남 서 북 (시계 방향)
dx = [0, 1, 0, -1] # x가 상하
dy = [1, 0, -1, 0] # y가 좌우

# 방향 전환하는 함수
def turn(direction, C):
    if C == 'L':
        direction = (direction - 1) % 4 # 위치 정보만 반환해주면 됨 => 모듈러 연산으로 나머지 값을 구함 => dx, dy에서 해당 위치에 맞게 결과 반환
    else:
        direction = (direction + 1)  % 4
        
    return direction
        
def solution():
    x, y = 1, 1         # 맨 처음 시작
    board[x][y] = 2      # 뱀이 존재하는 위치는 2로 표시 (굳이 2가 아니어도 됨)
    direction = 0       # 처음엔 동쪽을 보고 있음 (dx, dy에서 0이 동쪽)
    time = 0            # 시직한 뒤에 지난 '초' 시간
    index = 0           # 다음에 회전할 정보
    q = [(x, y)]        # 뱀이 차지하고 있는 위치 정보 (꼬리가 앞쪽) => 매 시점마다 뱀이 존재하는 위치를 항상 2차원 리스트에 기록
    
    while True:
        nx = x + dx[direction]
        ny = y + dy[direction]
        
        # 맵 범위 안에 있고, 뱀의 몸통이 없는 위치라면
        if 1 <= nx and nx <= N and 1 <= ny and ny <= N and board[nx][ny] != 2:
            
            # 사과가 없다면 이동 후에 꼬리 제거
            if board[nx][ny] == 0:
                board[nx][ny] = 2
                q.append((nx, ny))
                px, py = q.pop(0)
                board[px][py] = 0 # 다시 0으로 바꿈
                
            # 사과가 있다면 이동 후에 꼬리를 제거하지 않는다
            if board[nx][ny] == 1:
                board[nx][ny] = 2
                q.append((nx, ny))
                
        # 벽이나 뱀의 몸통과 부딪혔다면
        else:
            time += 1 # 부딪힐 때도 시간은 1초 증가하니까 계산해줘야함
            break
            
        # 다음 위치로 머리를 이동
        x, y = nx, ny
        time += 1
        
        # index : 회전할 정보 => 이게 L이랑 같아지면 더 이상 회전할 조건이 주어지지 않았다는 뜻
        if index < L and time == direc[index][0]: # direc[idnex][0] => 회전할 시간
            direction = turn(direction, direc[index][1])
            index += 1
    return time


print(solution())
```
